```python
# encoding=utf-8
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
import jieba
jieba.set_dictionary('dict.txt.big.txt')
from gensim import corpora, models, similarities
from collections import defaultdict
from pprint import pprint
import os
path = 'lyric/'
```

# import files

```python
documents = []
dict2lyric = []
# read all files from folder lyric and build documents
for i,filename in enumerate(os.listdir(path)):
    
    try:
        with open( path+filename , 'r', encoding='UTF-8') as fin:
            content = fin.read()
            #print(content)
            content = content.replace(' ', '')
            content = content.replace('\n','')
            #print(content)
            documents.append(content)
        dict2lyric.append(filename)
        print(i,filename)
    except:
        pass
# display documents
#pprint(documents)
#print(len(documents))
```

# 中文斷字處理

```python
texts = []
# remove common words and tokenize
stoplist = set('你 我 他 ( ) ， 你們 我們 他們 的 有'.split()) # set
for content in documents:
    #print(content)
    words = jieba.cut(content, cut_all=False)
    texts.append([word for word in words if word not in stoplist])
#pprint(texts)

# remove words that appear more than three times
frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token] += 1 
texts = [[token for token in text if frequency[token] > 1] for text in texts]
pprint(texts)
```

# Build dictionary and corpus

```python
dictionary = corpora.Dictionary(texts)
print(dictionary.token2id) # Reveal Mapping Result(words and their id)
dictionary.save('/tmp/deerwester.dict') # store the dictionary, for future reference
print(dictionary)

corpus = [dictionary.doc2bow(text) for text in texts]
corpora.MmCorpus.serialize('/tmp/deerwester.mm', corpus) # store to disk, for later use
print(corpus)
```

# Transformation

```python
# reload
dictionary = corpora.Dictionary.load('/tmp/deerwester.dict')
corpus = corpora.MmCorpus('/tmp/deerwester.mm')
# use this tiny corpus to define a 2-dimensional LSI space:
lsi = models.LsiModel(corpus, id2word=dictionary, num_topics=2)
```

# Similarity

```python
# Given a query
doc = '可樂'
query = []
# 中文斷字
doc = doc.replace(' ', '')
doc = doc.replace('\n','')
doc = jieba.cut(doc, cut_all=False)
query = [word for word in doc if word not in stoplist]
print(query)

vec_bow = dictionary.doc2bow(query)
vec_lsi = lsi[vec_bow] # convert the query to LSI space
print(vec_lsi)

# Initializing query structures
index = similarities.MatrixSimilarity(lsi[corpus]) # transform corpus to LSI space and index it

# Index persistency is handled via the standard save() and load() functions:
index.save('/tmp/deerwester.index')
index = similarities.MatrixSimilarity.load('/tmp/deerwester.index')
```

# Perform queries

```python
# obtain similarities of our query document against the nine indexed documents: (using cosine similarity)
sims = index[vec_lsi] # perform a similarity query against the corpus
print(list(enumerate(sims))) # print (document_number, document_similarity) 2-tuples

# Cosine measure returns similarities in the range <-1, 1> (the greater, the more similar), 
# Cos(0) = 1, 代表夾角很小 => 所以越接近
# so that the first document has a score of 0.99809301 etc.

# Display query result
sims = sorted(enumerate(sims), key=lambda item: -item[1])
print(sims) #結果最接近1,代表關聯度越高

# 前三大關聯結果
for i in range(3):
    print(dict2lyric[sims[i][0]])
```
